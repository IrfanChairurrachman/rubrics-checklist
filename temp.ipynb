# Score Calculation

```python
fnb = pd.read_csv('data-input/ml_fnb.csv')
fnb
```

```python
fnb_clean = fnb.ffill()
fnb_clean[['Category', 'Major Point']] = fnb_clean[['Category', 'Major Point']].astype('category')
```

```python
fnb_major = fnb_clean.groupby(['Major Point'], sort=False).max('Max Point').reset_index()
fnb_major
```

```python
fnb_cat = fnb_clean[['Category', 'Major Point']].drop_duplicates()
fnb_cat
```

```python
cat_dict = fnb_cat.groupby(['Category'], sort=False)['Major Point'].apply(list).to_dict()
for key in cat_dict:
    print(key)
    for i, value in enumerate(cat_dict[key]):
        print(value)
        print(i)
```

# Text Template

```python
text_output = ""
count = 0
earned = int(fnb_major.loc[count, 'Earned'])
max_point = int(fnb_major.loc[count, 'Max Point'])
icon = '✅' if earned == max_point else '❌'

for cat in cat_dict:
    text_output += cat
    for value in cat_dict[cat]:
        text_output += '\n' + f"{icon} [{earned}/{max_point}] {value}"
    text_output += '\n\n'

print(text_output)
```

```python
from string import Template

# read text from folder template_text, file example.txt
with open('template.txt', mode='r', encoding='utf-8') as f:
    content = f.read()
    temp = Template(content)
    example = temp.substitute(
        NAME = "Seseorang",
        CASE = "FnB",
        POINTS = text_output,
        TOTAL = int(fnb_major['Earned'].sum())
        )

print(example)
```
